# 1. Read data from excel sheet out_joined.csv and out_merge.csv into 2 different dataframes.

```python
import pandas as pd
from pandas import DataFrame

import sqlite3 as sql

df_joined = pd.read_csv(r'F:\\INTERNSHIP\\Week-2(python)\\data\\data sets\\Question_Set_2_Data\\out_joined.csv')
df_merge = pd.read_csv(r'F:\\INTERNSHIP\\Week-2(python)\\data\\data sets\\Question_Set_2_Data\\out_merge.csv')
```

# 2. Create a new database as 'population.db' and dump data from point 1 into 2 sqlite tables.

```python
#create database
conn = sql.connect('population.db')  
c = conn.cursor()

#create table
df_joined.to_sql('out_joined',conn,if_exists = 'replace',index = False)
df_merge.to_sql('out_merge',conn,if_exists = 'replace',index = False)

#show thorugh sqlite cursor
c.execute('''select * from out_joined''')

#storing in dataframe for more specific view and specifying column name 
dfjoined = DataFrame(c.fetchall(),columns = ['id','Indicator','Facility Type','Total No. of Facilities #',
                                       'Facilities reporting nil performance - Numbers*',
                                       'Details of Maximum and Minimum Performing facilities - Performance',
                                       'Details of Maximum and Minimum Performing facilities - Value Reported',
                                       'Details of Maximum and Minimum Performing facilities - Sub-district',
                                       'Details of Maximum and Minimum Performing facilities - Facility'])

c.execute('''select * from out_merge''')

dfmerge = DataFrame(c.fetchall(),columns = ['id','Indicator','Facility Type','Total No. of Facilities #',
                                        'Facilities reporting nil performance - Numbers*',
                                        'Details of Maximum and Minimum Performing facilities - Performance',
                                        'Details of Maximum and Minimum Performing facilities - Value Reported',
                                        'Details of Maximum and Minimum Performing facilities - Sub-district',
                                        'Details of Maximum and Minimum Performing facilities - Facility',
                                        'index','sepal_length','sepal_width','petal_length','petal_width','species'])
dfjoined.set_index('id',inplace = True)
dfmerge.set_index('id',inplace = True)
dfjoined.head(2)
```

# 6. Create a dataframe of
					"A": [12, 4, 5, None, 1],  
					"B": [7, 2, 54, 3, None],  
					"C": [20, 16, 11, 3, 8],  
					"D": [14, 3, None, 2, 6]} 
	And add value '20' to all coulmns.


```python
data = {
    "A": [12, 4, 5, None, 1],  
    "B": [7, 2, 54, 3, None],  
    "C": [20, 16, 11, 3, 8],  
    "D": [14, 3, None, 2, 6]
}

data_df = DataFrame(data)
data_df
```

```python

data_df['A'] = data_df.A+20
data_df['B'] = data_df.B+20
data_df['C'] = data_df.C+20
data_df['D'] = data_df.D+20

data_df
```

# 7. Remove/ Replace NAN values, If there exist in 'joined_df' and 'merged_df' dataframes.

```python
joined_df.isnull().sum()
```

```python
#merged_df = merged_df.dropna()

merged_df['Indicator'].fillna("no indicator",inplace = True)
merged_df['Facility Type'].fillna("no facility",inplace = True)
#merged_df['Total No. of Facilities #'].fillna(0,inplace = True)
merged_df['Total No. of Facilities #'].fillna(merged_df['Total No. of Facilities #'].mean(),inplace = True)
merged_df['Facilities reporting nil performance - Numbers*'].fillna(0,inplace = True)
merged_df['Details of Maximum and Minimum Performing facilities - Performance'].fillna('Neutral',inplace = True)
merged_df['Details of Maximum and Minimum Performing facilities - Value Reported'].fillna(0,inplace = True)
merged_df['Details of Maximum and Minimum Performing facilities - Sub-district'].fillna('no sub districts',inplace = True)
merged_df['Details of Maximum and Minimum Performing facilities - Facility'].fillna('no facilities',inplace = True)
merged_df.isnull().sum()
merged_df
```
